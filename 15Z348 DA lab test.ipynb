```python
df2 = df[['ge','tnp','twp','iap','fo','atd','esp']]

#the above attributes are the most descriptive, and hence they are used for the prediction
```

```python
import numpy as np
df2['ge'] = np.where(df2['ge']=='M',1,0)
df2['tnp'] = np.where(df2['tnp']=='Pass',1,
                      np.where(df2['tnp']=='Good',2,3)
                     )
df2['twp'] = np.where(df2['twp']=='Pass',1,
                      np.where(df2['twp']=='Good',2,3)
                     )
df2['iap'] = np.where(df2['iap']=='Pass',1,
                      np.where(df2['iap']=='Good',2,3)
                     )
df2['atd'] = np.where(df2['atd']=='Poor',1,
                      np.where(df2['atd']=='Average',2,3)
                     )
df2['fo'] = np.where(df2['fo']=='Business',1,
                      np.where(df2['fo']=='Farmer',2,
                               np.where(df2['fo']=='Others',3,
                                        np.where(df2['fo']=='Retired',4,5)
                                       )
                              )
                    )
```

```python
#using the above dataset, i will build a naive bayes classifier to predict the end semester percentage
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
import time

gnb = GaussianNB()
X_train, X_test = train_test_split(df2,test_size=0.25,random_state=int(time.time()))
```

```python
used_features =[
    "ge",
    "tnp",
    "twp",
    "iap",
    "fo",
    "atd"
]
gnb.fit(X_train[used_features].values,X_train["esp"])
y_pred = gnb.predict(X_test[used_features])

print("Number of mislabeled points out of a total {} points : {}, performance {:05.2f}%"
      .format(
          X_test.shape[0],
          (X_test["esp"] != y_pred).sum(),
          100*(1-(X_test["esp"] != y_pred).sum()/X_test.shape[0])
))
```
